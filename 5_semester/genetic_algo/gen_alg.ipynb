```python
import numpy as np
import matplotlib.pyplot as plt
import subprocess
import os, sys
%matplotlib inline
```

```python
def read(file):
    res = []
    with open(file, 'r' ) as f:
        i = 0
        for line in f:
            i+=1
            if i%2 == 0:
                continue
            line = line.split(' ')
            res.append(line[-2])
            
    return res
```

```python
def plot(N):
    res = read("stats.txt")
    plt.plot(range(1, 1000, 100), res, marker = 'o', linestyle = '-', label="%d"%N)
    plt.show()
    os.remove("stats.txt")
```

```python
os.remove("stats.txt")
for N in range(10000, 100000, 10000):
    for step in range(1, 1000, 100):
        subprocess.call(["./run", str(2), str(1000), str(step), "--file", "graph(1).txt"])
        print(step)
    plot(N)
```

Из графиков видно, что после $100$ итераций увеличивать это число бессмысленно, потому что функция может застрять в локальном минимуме. Из-за недетерминированности алгоритма, результаты флуктуируют.

$Crossover:$

Первая половина пути перется от первого потомка, вторая - по порядку от начала второго потомка, исключая совпадения с первой половиной.

$Selection:$

Все особи скрещиваются и мутируют, затем считается фиттинг-функция. Скрещивается $80$% от общего числа, а мутирует -- $20$%

$Mutation$:

Особь мутирует с вероятностью $20$%, рандомно выбирается точка пути, которая поменяется с правой местами.
